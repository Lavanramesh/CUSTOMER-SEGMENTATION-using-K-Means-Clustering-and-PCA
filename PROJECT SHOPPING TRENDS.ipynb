```python
category_summary=df.groupby('Category')['Purchase_Amount_USD'].agg(['count', 'sum', 'mean']).sort_values(by='count')
```

```python
# Plot Purchase Count by Category
plt.figure(figsize=(5,6))
colors=['red' if x < 500 else 'blue' for x in category_summary['count']]  
sns.barplot(x=category_summary.index, y=category_summary['count'],palette=colors)
plt.title('Purchase Count by Product Category')
plt.show()
```

```python
# Step 1: Create age groups
df['Age_Group'] = pd.cut(df['Age'],
                         bins=[17, 25, 35, 50, 100],
                         labels=['18-25', '26-35', '36-50', '50+'])

# Step 2: Standardize payment method values
df['Payment_Method'] = df['Payment_Method'].astype(str).str.strip().str.lower()

# Step 3: Filter only credit card payments
card_df = df[df['Payment_Method'] == 'credit card']

# Step 4: Group by Age Group
age_spending = card_df.groupby('Age_Group')['Purchase_Amount_USD'].sum()

plt.figure(figsize=(8, 6))
age_spending.plot(kind='bar', color='skyblue')
plt.title('Credit Card Spending by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Purchase Amount (USD)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
df['Season'] = df['Season'].astype(str).str.strip().str.lower()
season_spending = card_df.groupby('Season')['Purchase_Amount_USD'].sum()

plt.figure(figsize=(8, 6))
season_spending.plot(kind='bar', color='coral')
plt.title('Credit Card Spending by Season')
plt.xlabel('Season')
plt.ylabel('Total Purchase Amount (USD)')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
location_spending = card_df.groupby('Location')['Purchase_Amount_USD'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 8))
location_spending.plot(kind='barh', color='lightgreen')
plt.title('Credit Card Spending by Location')
plt.xlabel('Total Purchase Amount (USD)')
plt.ylabel('Location')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns


# Step 1: Data Cleaning
df['Discount_Applied'] = df['Discount_Applied'].str.strip().str.lower().map({'yes': 1, 'no': 0})
df['Subscription_Status'] = df['Subscription_Status'].str.strip().str.lower().map({'yes': 1, 'no': 0})
df['Promo_Code_Used'] = df['Promo_Code_Used'].str.strip().str.lower().map({'yes': 1, 'no': 0})
df['Gender'] = df['Gender'].str.strip().str.lower().map({'male': 1, 'female': 0})

# Step 2: Select features for clustering
features = df[['Age', 'Purchase_Amount_USD', 'Previous_Purchases', 'Review_Rating',
               'Subscription_Status', 'Discount_Applied', 'Promo_Code_Used', 'Gender']]

# Drop any missing values
features = features.dropna()

# Step 3: Normalize the data
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Step 4: Apply K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(features_scaled)

# Add cluster labels back to the dataframe
df['Cluster'] = clusters

# Step 5: Analyze Cluster Profiles
cluster_summary = df.groupby('Cluster')[['Age', 'Purchase_Amount_USD', 'Previous_Purchases',
                                         'Review_Rating', 'Subscription_Status',
                                         'Discount_Applied', 'Promo_Code_Used']].mean()

print("\n Cluster Profile Summary:")
print(cluster_summary)

# Step 6: Visualize Clusters using PCA
pca = PCA(n_components=2)
components = pca.fit_transform(features_scaled)

plt.figure(figsize=(8, 6))
sns.scatterplot(x=components[:, 0], y=components[:, 1], hue=clusters, palette='Set2', s=100)
plt.title('Customer Clusters (PCA Projection)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.tight_layout()
plt.show()
```
